## Carga de datos Python

```python
# Definición de funciones para leer archivos csv
import csv

def crearMatrices():
    data = []
    with open('avocado.csv') as File:  
        reader = csv.reader(File)
        for row in reader:
            data.append(row)
    return data

def totalValores():
    with open('avocado.csv', newline='') as File:  
        reader = csv.reader(File)
        contador = 0
        for row in reader:
            contador += 1
    print("Exiten ", contador, " objetos en la lista")

def minimo(m):
    min = m[1][2]
    for i in range(2, len(m)):
        if m[i][2] < min :
            min = m[i][2]
    return min

def maximo(m):
    max = m[1][2]
    for i in range(2, len(m)):
        if m[i][2] > max :
            max = m[i][2]
    return max
```
